```python
import  graph_tool.all as gt
```

```python
N=15
g = gt.Graph(directed = False)
g.add_vertex(N)
```

```python
adj_list={0:[1,2],
          1:[2,3,14],
          2:[3,7],
          3:[4],
          4:[5,6,7],
          5:[6,8,9],
          6:[7,8],
          7:[8,9],
          8:[9],
          9:[10],
          10:[11,12,14],
          11:[12,13],
          12:[14],
          13:[14],
          }

for v,es in adj_list.items():
    for v_ in es:
        g.add_edge(v, v_)
```

```python
pos=gt.sfdp_layout(g, cooling_step=0.95, epsilon=1e-2)
gt.graph_draw(g, vertex_text=g.vertex_index,pos=pos, output_size=(200,200))
```
